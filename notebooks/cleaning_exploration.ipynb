```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('production_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Add Good_Chips and Yield columns
df['Good_Chips'] = df['Total_Chips_Produced'] - df['Defective_Chips']
df['Yield_%'] = (df['Good_Chips'] / df['Total_Chips_Produced']) * 100
```

```python
# Missing values and types
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Plot production over time
plt.figure(figsize=(10, 4))
plt.plot(df['Date'], df['Total_Chips_Produced'], label='Total')
plt.plot(df['Date'], df['Good_Chips'], label='Good')
plt.title("Chip Production Over Time")
plt.legend()
plt.grid()
plt.show()
```

```python
# Yield distribution
sns.histplot(df['Yield_%'], bins=20, kde=True, color='green')
plt.title("Yield Distribution (%)")
plt.show()
```
