```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('equipment_logs.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Most frequent errors
top_errors = df['Error_Code'].value_counts().head(5)
print("Top 5 Error Codes:\n", top_errors)
```

```python
# Downtime per machine
downtime = df.groupby('Machine_ID')['Downtime_Minutes'].sum().sort_values()
downtime.plot(kind='barh', color='orangered')
plt.title("Total Downtime per Machine")
plt.xlabel("Minutes")
plt.show()
```

```python
# Daily downtime trend
daily_downtime = df.groupby('Date')['Downtime_Minutes'].sum()
daily_downtime.plot(figsize=(10,4), title="Daily Total Downtime", grid=True)
plt.show()
```
